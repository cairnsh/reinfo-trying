A Jupyter notebook with some tools for plotting recordings.

```python
# Hide some of the code in a module, view_recording_utils.py,
# to make the notebook less cluttered.

import numpy as np

from view_recording_utils import load_the_recording, \
    memory, calculate_reward_tables, calculate_rewards_per_agent

nplayers, recording = load_the_recording()

print("Number of players:", nplayers)

EPISODE_LENGTH = 100
```

`nplayers` is the number of players in the recording.

`recording` is a series of tuples of the form `(permutation, actions)`
- `permutation` records the matching between players. Player $j$ is matched with player ${\tt permutation}(j)$.
- `actions` records the action of the players. Typically 0 = cooperate and 1 = defect.

The number of games per episode is not recorded, but it's always been 100 so far.

```python
reward = {
    (0, 0): 0,
    (0, 1): 1,
    (1, 0): 2,
    (1, 1): 0
}
```

These are the game's rewards. Make sure that they match the ones in `train.py`!

```python
# Rearrange the output of calculate_reward_tables into a history
# of average rewards for a specific player pair

two_player_histories = {
    (i, j): []
    for i in range(nplayers) for j in range(nplayers) if i != j
}

for reward_table_at_timestep in calculate_reward_tables(reward, nplayers, recording):
    for playerpair, averagereward in reward_table_at_timestep.items():
        two_player_histories[playerpair].append(averagereward)
```

```python
# Plot the histories of two-player rewards

from matplotlib import pyplot as p

figure, boxes = p.subplots(nplayers, nplayers, figsize=(24, 12))

figure.set_facecolor((1, 1, 1))
figure.set_alpha(1)

def plotscale(box, data):
    box.plot([j * 8000 for j in range(len(data))], data)
    
for i in range(nplayers):
    for j in range(nplayers):
        if i == j:
            boxes[i, j].axis('off')
            continue
        boxes[i, j].set_ylim([0, 2])
        
        plotscale(boxes[i, j], two_player_histories[i, j])
        plotscale(boxes[i, j], two_player_histories[j, i])

p.show()
```

```python
# Plot the histories of single-player rewards

one_player_histories = [[] for j in range(nplayers)]

for re in calculate_rewards_per_agent(reward, nplayers, recording):
    for player in range(nplayers):
        one_player_histories[player].append(re[player])

figure, boxes = p.subplots(6, 1, figsize=(24, 24))

for j in range(nplayers):
    boxes[j].set_ylim([0, 2])
    plotscale(boxes[j], one_player_histories[j])

p.show()
```

```python
# Plot the history of one versus five

figure, box = p.subplots(figsize=(24, 8))

box.set_xlim([700000, 720000])
plotscale(box, two_player_histories[0, 4])
plotscale(box, two_player_histories[4, 0])
p.show()
```

```python
one_versus_five_games = []
i = 0
for permutation,actions in recording:
    i += 1
    if permutation[0] == 4: # one vs five
        one_versus_five_games.append((i, actions[0], actions[4]))

one_versus_five_episodes = [
    one_versus_five_games[idx:idx+EPISODE_LENGTH]
    for idx in range(0, len(one_versus_five_games), 100)
]

one_versus_five_means = np.array([np.mean(z, axis=0) 
                         for z in one_versus_five_episodes])
```

```python
def calculate_reward_of_episode(episode):
    total = 0
    for line in episode:
        _,a,b = line
        total += reward[a, b]
        total += reward[b, a]
    return total

def get_episode_start_timestep(episode):
    return episode[0][0]

p.scatter([get_episode_start_timestep(episode) for episode in one_versus_five_episodes],
          [calculate_reward_of_episode(episode) for episode in one_versus_five_episodes])
p.show()
```
